Q2. What is the difference between Hard Link & Soft Link ?

Ans: A hard link is a reference that points directly to the inode (data structure) of a file on the disk. It creates another filename for the same data, and both the original file and the hard link are treated equally by the filesystem.

`Command: ln source_file hard_link_name`

A soft link (or symbolic link) is a special file that contains a path to another file or directory. It acts like a shortcut and refers to the file name rather than the data itself.

`Command: ln -s source_file soft_link_name`

Q3. What is a Kernel in Linux ?

Ans: In Linux the kernel is the core component of the operating system. It acts as a bridge between hardware and software, managing system resources such as CPU, memory, and device drivers. Linux Kernel is like the brain of the operating system .Here are a few key roles of the Linux kernel:


*  Process Management: Handles multitasking, scheduling, and execution of
 processes.

*   Memory Management: Allocates and deallocates memory efficiently for applications.

*   Device Management: Communicates with hardware components through device drivers.

*   File System Management: Provides access to various file systems and storage devices.

*   Networking: Manages network communications and protocols.

It will ask the password to create the group.

 Step 2: Create and add user p1 to groups g1, g2, and g3
*    Command:
```
sudo groupadd g1
sudo groupadd g2
sudo groupadd g3
```

```
# Add user p1 to all three groups
 Command: sudo usermod -aG g1,g2,g3 p1
```

```
# Step 3: Make g1 the default group for user p1
 command:
` sudo usermod -g g1 p1`

# to check user detail
 * command:
` id p1`

# To confirm g1 is the default group:
 command:
` grep p1 /etc/passwd`
```
