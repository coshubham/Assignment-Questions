3.  starting from e feature branch, push the branch to the master, following the architecture

```python
# 1. Initialize repo and commit base file to master
git init
touch main.txt
git add main.txt
git commit -m "Initial commit"

# 2. Create develop branch
git checkout -b develop

# 3. Create feature branch from develop
git checkout -b feature/login
touch login.txt
git add login.txt
git commit -m "Add login feature"

# 4. Merge feature to develop
git checkout develop
git merge feature/login
```

4. Push an urgent.txt on master using hotfix

```python
# Start from master
git checkout master

# Create hotfix branch
git checkout -b hotfix

# Create urgent file
touch urgent.txt
git add urgent.txt
git commit -m "Add urgent.txt as hotfix"

# Merge hotfix to master
git checkout master
git merge hotfix
```
